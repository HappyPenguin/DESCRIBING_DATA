## Data Visualisation and Exploration in Multiple Regression Analyses

Let's start by looking at a very simple linear regression.
Does height increase with age?

```python
import numpy as np
import matplotlib.pylab as plt

# Start by creating some data
age = np.random.random(100) * 5 + 6
h_noise = np.random.random(100)
height = age * 12 + h_noise * 25
height_name = 'Height'
age_name = 'Age'

# Set a few constants for all the plots:
figsize = (10,6)
dpi = 80
axis_fontsize = 25
title_fontsize = 25

# Now plot age against height
def simple_x_y (x, y, x_name, y_name, title):
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    
    ax.scatter(x, y)
    ax.set_title(title, fontsize=title_fontsize)
    ax.set_ylabel(y_name, fontsize=axis_fontsize)
    ax.set_xlabel(x_name, fontsize=axis_fontsize)

title = 'Height increases with age'
simple_x_y(age, height, age_name, height_name, title)
```

Well, yes it does. And it looks like a pretty strong correlation..but how strong?

```python
import sys
sys.path.insert(0,'/home/kw401/CAMBRIDGE_SCRIPTS/PYTHON_SCRIPTS/')
import ols

ols_height_age = ols.ols(height, age, y_varnm=height_name, x_varnm=[age_name])
ols_height_age.summary()
```

Summary gives you all the information in one place.

One of the outputs from running ordinary least squares regression is the residual variance. It's called m.e.

height = residuals + estimated height

```python
res_height_age = ols_height_age.e
res_height_age_name = 'Residuals'
est_height = height - res_height_age

def regline_x_y(x, y, est_y, x_name, y_name, title):
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    ax.scatter(x, y)
    ax.plot(x, est_y, color='r')
    ax.set_title(title, fontsize=title_fontsize)
    ax.set_ylabel(y_name, fontsize=axis_fontsize)
    ax.set_xlabel(x_name, fontsize=axis_fontsize)

title = ('Height increases with age')
regline_x_y(age, height, est_height, age_name, height_name, title)
```

```python
def regline_res_x_y(x, y, est_y, x_name, y_name, title):
    fig = plt.figure(figsize=figsize, dpi=dpi)
    ax = fig.add_subplot(111)
    ax.scatter(x, y)
    ax.plot(x, est_y, color='r')
    for i in range(0,len(x)):
        ax.plot(np.hstack([x[i], x[i]]),
                np.hstack([y[i], est_y[i]]),
                color='g')
    ax.set_title(title, fontsize=title_fontsize)
    ax.set_ylabel(y_name, fontsize=axis_fontsize)
    ax.set_xlabel(x_name, fontsize=axis_fontsize)

title = ('Height increases with age')
regline_res_x_y(age, height, est_height, age_name, height_name, title)
```

The length of the green lines will tell us how far each person lies from the regression line

```python
res_name = 'Residuals'

ols_res_height_age_age = ols.ols(res_height_age, age, y_varnm=res_name, x_varnm=[age_name])
ols_res_height_age_age.summary()

res_height_age_age_res = ols_res_height_age_age.e
est_res_height_age = res_height_age - res_height_age_age_res

title = 'The residuals do not correlate with age'
regline_x_y(age, res_height_age, est_res_height_age, age_name, res_name, title)
```

Now lets create some more data. What if gender played a part in the relationship between age and height? what about the person's weight?

Now everything starts to get a bit more complicated. We can replicate the graphs that we've made so far though with some real data downloaded from: http://www.kovcomp.co.uk/support/XL-Tut/multiple-linear-regression-mlr.html

From that website:
"The data have been obtained in Lewis T. and Taylor L.R. (1967). Introduction to Experimental Ecology, New York: Academic Press, Inc.. They concern 237 children, described by their gender, age in months, height in inches (1 inch = 2.54 cm), and weight in pounds (1 pound = 0.45 kg)."

```python
# Load in some real data
data = np.genfromtxt('/home/kw401/DATA_FROM_WEB/Male_Age_Height_Weight.csv',
                        delimiter=',',
                        names=True)

age_name = 'Age'
height_name = 'Height'
male_name = 'Male'
weight_name = 'Weight'
age = data[age_name]
height = data[height_name]
weight = data[weight_name]
male = data[male_name]

# Age is given in months but I think the numbers aren't intuitive
# so change the units to years
age = age/12

# Replicate the analyses we've run so far
title = 'Height increases with age'
simple_x_y(age, height, age_name, height_name, title)

ols_height_age = ols.ols(height, age, y_varnm=height_name, x_varnm=[age_name])
ols_height_age.summary()

res_height_age = ols_height_age.e
est_height = height - res_height_age

title = 'Height increases with age'
regline_res_x_y(age, height, est_height, age_name, height_name, title)
```

```python
res_name = 'Residuals'

ols_res_height_age_age = ols.ols(res_height_age, age, y_varnm=res_name, x_varnm=[age_name])
ols_res_height_age_age.summary()

res_height_age_age_res = ols_res_height_age_age.e
est_res_height_age = res_height_age - res_height_age_age_res

title = 'The residuals do not correlate with age'
regline_x_y(age, res_height_age, est_res_height_age, age_name, res_name, title)
```

It's pretty reasonable to suspect that as you get older you not only get taller but weigh more too.

Let's take a look:

```python
ols_weight_age = ols.ols(weight, age, x_varnm=[age_name], y_varnm=weight_name)
ols_weight_age.summary()

res_weight_age = ols_weight_age.e
est_weight = weight - res_weight_age

title = 'Weight increases with age'
regline_x_y(age, weight, est_weight, age_name, weight_name, title)
```

But, oh my goodness, as you get taller you also weigh more. So that has to be related, right?!

```python
ols_weight_height = ols.ols(weight, height, y_varnm=weight_name, x_varnm = [height_name])

res_weight_height = ols_weight_height.e
est_weight_height = weight - res_weight_height

'Weight increses with height'
regline_x_y(height, weight, est_weight_height, height_name, weight_name, title)
```

So now the question comes: which relationship is the most important? Is there really a relationship between height and weight or are they just correlated because they are both correlated with age?

We can perform a multiple regression to see what might be happening in the data

```python
ols_weight_height_age = ols.ols(weight, np.vstack([height, age]).T, y_varnm=weight_name, x_varnm = [height_name, age_name])
ols_weight_height_age.summary()
```

We would interpret this table as:

"There is a significant relationship between height and weight, controlling for the relationships between age and height and age and weight.  There is also a significant relationship between age and weight, controlling for the relationships between age and height and height and weight."

This would usually just be stated as "controlling for the effects of age" (or height - depending on which relationship you were interested in).

### Added Variable Plots

So given that these three variables are all related, the challenge becomes: how do we graphically investigate or present this data?

We can look at all three of the pairwise plots we have created already (which is important) and also create added variable plots.

Lets start by looking at the relationship between height and weight controlling for age.

You can think of this as a correlation of the residual variance after correlating height with age, and the residual variance after correlating weight with age.

```python
res_height_age_name = 'Residual variance in height after correlating with age'
res_weight_age_name = 'Residual variance in weight after correlating with age'
axis_fontsize=12

title = 'Added variable plot for height vs weight'
simple_x_y(res_height_age, res_weight_age, res_height_age_name, res_weight_age_name, title)
```

Definitely looks like there is a relationship there, lets take a look at a correlation between these two sets of residuals:

```python
res_height_age_name = 'Height | age'
res_weight_age_name = 'Weight | age'

ols_res_res_weight_height_age = ols.ols(res_weight_age, res_height_age, x_varnm=[res_height_age_name], y_varnm= res_weight_age_name)

ols_res_res_weight_height_age.summary()

ols_weight_height_age.summary()
```

Note that the coefficent for height in the multiple regression is exactly the same as when we correlate the residuals (because that's how it's calculated!)

We can replicate the same effect for age against weight:

```python
ols_weight_height = ols.ols(weight, height, y_varnm=weight_name, x_varnm=[height_name])
res_weight_height = ols_weight_height.e
res_weight_height_name = 'Weight | height'

ols_age_height = ols.ols(age, height, y_varnm=age_name, x_varnm=[height_name])
res_age_height = ols_age_height.e
res_age_height_name = 'Age | height'

ols_res_res_weight_age_height = ols.ols(res_weight_height, res_age_height, x_varnm=[res_age_height_name], y_varnm= res_weight_age_name)
res_res_weight_age_height = ols_res_res_weight_age_height.e

est_res_weight_height = res_weight_height - res_res_weight_age_height

axis_fontsize=12
title = 'Added variable plot for age vs weight'
regline_x_y(res_age_height, res_weight_height, est_res_weight_height, res_age_height_name, res_weight_height_name, title)

ols_res_res_weight_age_height.summary()
```
